```python
from openai import OpenAI

client = OpenAI(
    api_key=""
)
```

```python
def generate(messages):
    response = client.chat.completions.create(
        model="gpt-4o-2024-08-06", messages=messages, temperature=0.6, logprobs=True
    )
    return response
```

```python
import numpy as np

TAU_THRESHOLD = 50

def get_tau(logprobs):
    tau = 0
    for logprob in logprobs:
        tau += logprob.logprob
    tau /= len(logprobs)
    print(tau)
    tau = np.round(np.exp(tau) * 100, 2)
    return tau
```

```python
for i in range(1):
    dialogue = example_data["dialogue"]
    messages = [
        {
            "role": "system",
            "content": f'You are in role-play, playing {example_data["meta"]["name"]}\n{example_data["meta"]["profile"][example_data["meta"]["name"]]}',
        }
    ]
    for message in dialogue:
        if message["from"] == "User":
            print("USER:", message["value"])
            messages.append({"role": "user", "content": message["value"]})
            response = generate(messages)
            print("BOT:", response.choices[0].message.content)
            tau = get_tau(response.choices[0].logprobs.content)
            print(tau, "%")
            if tau < TAU_THRESHOLD:
                messages = messages[:-1]
                print("SYSTEM: REJECT RESPONSE AND REMOVE USER MESSAGE DUE TO LOW TAU")
            else:
                messages.append(
                    {
                        "role": "assistant",
                        "content": response.choices[0].message.content,
                    }
                )
```
